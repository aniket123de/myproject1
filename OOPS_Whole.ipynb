## **Class Variable**

````python
class Student:
  classVariable = 10
  def display(self):
    self.objectVariable = 20
    print("Hello World")

res = Student()
res.display()
print(res.classVariable)
print(res.objectVariable)

ans = Student()
ans.display()
print(ans.classVariable)
print(ans.objectVariable)

print(Student.classVariable)
print(Student.objectVariable)
````




*   **classVariable -**
  *   can be accessed by Class and Object.
  *   self is not required


*   **objectVariable -**
  *   can be accessed by objects only.
  *   self is required





## **Constructor**

### **1. Default Constructor**

````python
class Demo:
  def __init__(self):
    self.a = 10
    self.b = 20

  def display(self):
    print(self.a + self.b)

res = Demo()
res.display()
````

*When we are creating an object "res" in Python , we call a constructor.*

*In the above program , this is default constructor , where we have set the value of 2 object variables 'a' and 'b' as 10 and 20.*

*At the time of object creation, the constructor of the class will be invoked automatically.*


---

So , here when we are defining , res = Demo()


---
Then the constructor of the class __init__
(self) will be invoked automatically.



### **2. Parameterised Constructor**

````python
class Order:
  def __init__(self,a,b):
    self.kabab = a
    self.nun = b

  def bill(self):
    print(self.kabab*120 + self.nun*40)


cust1 = Order(3,8)
cust1.bill()

cust2 = Order(2,15)
cust2.bill()
````
